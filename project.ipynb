#### Exercise 11. Analyse the relation between the surface and the price of the houses (★★☆)

Hint: You can make a `scatter plot` and then write a conclusion about it

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')

ds = ds[ds['price'] > 0]
ds = ds[ds['surface'] > 0]
ds = ds[ds['surface'] < 1000]
ds = ds[ds['price'].notna()]
ds = ds[ds['surface'].notna()]

x = ds['surface']
y = ds['price']

print(len(x) == len(y))

plt.figure(figsize = (10, 5))

plt.scatter(x, y, label = "Price")

plt.title("Surface VS Price")
plt.xlim([min(x), max(x)])
plt.ylim([min(y), max(y)])
plt.legend()
plt.show() 
```

**Analysis:** If we take the outliers out and scale the plot for houses below 1000m^2 in surface, we can see a very linear relationship between the surface area and the top price houses are sold. The minimum price a house is sold for seems to be independent of the surface area.

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

def absolute_maximum_scale(series):
    return series / series.abs().max()

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]

ds = ds[ds['level5'].isin(populations)]

plt.figure(figsize = (10, 5))

for population in populations:
    data = ds.loc[ds['level5'] == population, 'price']
    plt.hist(absolute_maximum_scale(data), bins = 30, alpha = 0.7, label=population)

plt.title("South Madrid Housing Prices")
plt.legend()
plt.show()


```

**Analysis:** While each population seems to have its own particularities, most of the houses seem to be within the same price range, which is between 20% and 40% of the most expensive houses in the set.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import pandas as pd
import matplotlib.pyplot as plt

def absolute_maximum_scale(series):
    return series / series.abs().max()

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]

ds = ds[ds['level5'].isin(populations)]

fig, axs = plt.subplots(len(populations), figsize=(10, 8))
fig.suptitle('South Madrid Housing Prices')
i=0

for population in populations:
    data = ds.loc[ds['level5'] == population, 'price']
    axs[i].plot(absolute_maximum_scale(data), alpha = 0.7, label=population)
    axs[i].set_title(population)
    i += 1

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon, LayerGroup

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Stamen.Terrain)

def absolute_maximum_scale(series):
    return series / series.abs().max()

def getIcon(color):
    return AwesomeIcon(
                name='home',
                marker_color=color,
                icon_color='black',
                spin=False
        )

def replaceCommas(coordinate):
    return float(coordinate.replace(',','.'))

ds = pd.read_csv('assets/real_estate.csv', sep=';')

populations = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
iconColors = ['green','red','blue','orange']

ds = ds[ds['level5'].isin(populations)][['level5','latitude','longitude']]
ds = ds[ds.notna().any(axis=1)]
markers = list()
i = 0

for population in populations:    
    reducedSet = ds[ds['level5'] == population]
    coordinates = zip(
        reducedSet['latitude'], reducedSet['longitude']
        )
    for tp in tuple(coordinates):
        markers.append(
            Marker(
                location=(replaceCommas(tp[0]), replaceCommas(tp[1])), 
                icon=getIcon(iconColors[i])
            )
        )
    i += 1

markerCluster = LayerGroup(layers=markers)

map.add_layer(markerCluster)

map


```
